```python
#perfect number
def is_perfect(num):
    if num < 1:
        return False
    divisors_sum = 0
    for i in range(1, num):
        if num % i == 0:
            divisors_sum += i
    return divisors_sum == num

start = int(input("Start of range: "))
end = int(input("End of range: "))

print(f"Perfect numbers between {start} and {end}:")
for number in range(start, end + 1):
    if is_perfect(number):
        print(number)
```

```python
#amicable
def sum_proper_divisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

def are_amicable(a, b):
    return sum_proper_divisors(a) == b and sum_proper_divisors(b) == a

# Input two numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if are_amicable(num1, num2):
    print(f"{num1} and {num2} are an amicable pair!")
else:
    print(f"{num1} and {num2} are NOT an amicable pair.")
```

```python
#balanced_number
def balanced_number(num_str):
    length = len(num_str)
    mid = length // 2

    # For odd length, exclude the middle digit
    left_part = num_str[:mid]
    right_part = num_str[-mid:]

    left_sum = sum(int(d) for d in left_part)
    right_sum = sum(int(d) for d in right_part)

    return left_sum == right_sum

digits = input("Enter the number: ")

if balanced_number(digits):
    print(f"{digits} is a balanced number.")
else:
    print(f"{digits} is not a balanced number.")
```

```python
def capitalize_after_space(text):
    result = ""
    capitalize_next = True  # Capitalize first letter too

    for char in text:
        if capitalize_next and char != " ":
            result += char.upper()
            capitalize_next = False
        else:
            result += char
            capitalize_next = (char == " ")

    return result

text = "hello world python"
print(capitalize_after_space(text))  # Output: Hello World Python
```

```python
class Bankaccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount >= 0:
            self.balance += amount
            print("Total balance:", self.balance)
        else:
            print("Invalid amount for deposit.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        elif amount >= 0:
            self.balance -= amount
            print("Remaining balance:", self.balance)
        else:
            print("Invalid amount for withdrawal.")

    def get_balance(self):
        return self.balance

# create account
account = Bankaccount("suman", 5000)
print("Account holder name:", account.account_holder)
print("Initial balance:", account.get_balance())

account.deposit(2000)
account.withdraw(1000)
print("Final balance:", account.get_balance())
```

```python
# Participant class following encapsulation principles
class Participant:
    def __init__(self):
        self.__participant_id = None
        self.__name = None
        self.__base_fee = 200.0
        self.__num_events = 0
        self.__event_charge_per_event = 100.0

    # Setter methods
    def set_participant_id(self, pid):
        self.__participant_id = pid

    def set_name(self, name):
        self.__name = name

    def set_num_events(self, num_events):
        if num_events >= 0:
            self.__num_events = num_events
        else:
            print("Number of events can't be negative.")

    # Getter methods
    def get_participant_id(self):
        return self.__participant_id

    def get_name(self):
        return self.__name

    def get_base_fee(self):
        return self.__base_fee

    def get_num_events(self):
        return self.__num_events

    def get_event_charge_per_event(self):
        return self.__event_charge_per_event

    def calculate_total_fee(self):
        return self.__base_fee + (self.__num_events * self.__event_charge_per_event)

    # Display all participant details
    def display_details(self):
        print("\nParticipant Record")
        print(self.__participant_id)
        print("Name:", self.__name)
        print(f"Base Registration Fee: {self.__base_fee}")
        print(f"Event Charges: {self.__num_events} events @ {self.__event_charge_per_event}/event")
        print(f"Total Fee: {self.calculate_total_fee()}")


# Main program (modular and readable)
def main():
    # Create instance
    participant = Participant()

    # Set values using setter methods
    participant.set_participant_id("201501")
    participant.set_name("Kane Rahsil")
    participant.set_num_events(3)

    # Display all details
    participant.display_details()


# Run the main function
if __name__ == "__main__":
    main()
```

```python
class FoodItem:
  def food_information(self,item_Id,item_name,price,quality):
    self.item_Id=item_Id
    self.item_name=item_name
    self.price=price
    self.quality=quality

  def calculate_total_bill(self,quantity):
    return self.price * quantity

  def packing_charge(self):
    return 20

  def delivery_charge(self,quantity):
    return quantity * 20

  def calculate_gst(self, total_bill, gst_rate=5):
    return total_bill * (gst_rate / 100)

  def display_bill(self, quantity, discount_percentage=0):
    total_bill = self.calculate_total_bill(quantity)
    discount_amount = total_bill * (discount_percentage / 100)
    gst_amount = self.calculate_gst(total_bill)
    packing_bill = self.packing_charge()
    delivery_bill = self.delivery_charge(quantity)
    final_bill = total_bill - discount_amount + gst_amount + packing_bill + delivery_bill

    print(f"item_Id:{self.item_Id}")
    print(f"item_name:{self.item_name}")
    print(f"quality:{self.quality}")
    print(f"price:{self.price}")
    print(f"quantity:{quantity}")
    print(f"total_bill:{total_bill}")
    print(f"discount_applied: {discount_percentage}%")
    print(f"discount_amount:{discount_amount}")
    print(f"delivery_charges:{delivery_bill}")
    print(f"gst_amount:{gst_amount}")
    print(f"packing_bill:{packing_bill}")
    print(f"final_bill:{final_bill}")

item1 = FoodItem()
item1.food_information(1, "Burger", 500, "Excellent")

item2 = FoodItem()
item2.food_information(2, "pizza", 500, "good")

item1.display_bill(1, 10) # 2 quantities, 10% discount
print("\n")
item2.display_bill(2,) # 2 quantities, no discount
```

```python
# Base class
class Car:
    def start_engine(self):
        pass

    def shift_gear(self):
        pass

    def acceleration_engine(self):
        pass

    def break_car(self):
        pass

    def drive_car(self):
        self.start_engine()
        self.shift_gear()
        self.acceleration_engine()
        self.break_car()


# Subclass for Tesla
class TeslaElectric(Car):
    def start_engine(self):
        print("Tesla engine started silently")

    def shift_gear(self):
        print("Tesla shift gear")

    def acceleration_engine(self):
        print("Tesla accelerating smoothly")

    def break_car(self):
        print("Tesla regenerative braking activated")


# Subclass for BMW
class BMW(Car):
    def start_engine(self):
        print("BMW engine started with a roar")

    def shift_gear(self):
        print("BMW shift gear")

    def acceleration_engine(self):
        print("BMW accelerating powerfully")

    def break_car(self):
        print("BMW hydraulic brakes applied")


# Usage
tesla = TeslaElectric()
bmw = BMW()

print("Driving Tesla:")
tesla.drive_car()

print("\nDriving BMW:")
bmw.drive_car()
```

```python
# Student class with encapsulation
class Student:
    def __init__(self, roll, name, age):
        self._roll = roll
        self._name = name
        self._age = age

    def get_roll(self):
        return self._roll

    def get_name(self):
        return self._name

    def get_age(self):
        return self._age

    def set_name(self, name):
        self._name = name

    def set_age(self, age):
        self._age = age

    def display_info(self):
        print(f"Roll No: {self._roll}, Name: {self._name}, Age: {self._age}")


# Manager class for handling students
class StudentManager:
    def __init__(self):
        self.students = {}

    def insert_student(self, roll, name, age):
        if roll in self.students:
            print("⚠️ Student with this roll number already exists.")
        else:
            self.students[roll] = Student(roll, name, age)
            print("✅ Student added successfully.")

    def update_student(self, roll, name=None, age=None):
        if roll in self.students:
            if name:
                self.students[roll].set_name(name)
            if age:
                self.students[roll].set_age(age)
            print("✅ Student updated successfully.")
        else:
            print("❌ Student not found.")

    def delete_student(self, roll):
        if roll in self.students:
            del self.students[roll]
            print("🗑️ Student deleted successfully.")
        else:
            print("❌ Student not found.")

    def display_all_students(self):
        if not self.students:
            print("📭 No students to display.")
        else:
            print("\n📋 All Student Records:")
            for student in self.students.values():
                student.display_info()
            print()


# Main console loop
def main():
    manager = StudentManager()

    while True:
        print("\n===== Student Management System =====")
        print("1. Insert Student")
        print("2. Update Student")
        print("3. Delete Student")
        print("4. Display All Students")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            roll = input("Enter Roll No: ")
            name = input("Enter Name: ")
            try:
                age = int(input("Enter Age: "))
                manager.insert_student(roll, name, age)
            except ValueError:
                print("❌ Age must be a number.")

        elif choice == '2':
            roll = input("Enter Roll No to update: ")
            name = input("Enter new name (leave blank to skip): ")
            age_input = input("Enter new age (leave blank to skip): ")
            age = int(age_input) if age_input.strip() else None
            manager.update_student(roll, name if name else None, age)

        elif choice == '3':
            roll = input("Enter Roll No to delete: ")
            manager.delete_student(roll)

        elif choice == '4':
            manager.display_all_students()

        elif choice == '5':
            print("👋 Exiting system. Goodbye!")
            break

        else:
            print("❌ Invalid choice. Try again.")


if __name__ == "__main__":
    main()
```

```python
class Movie:
    def __init__(self, movieId, title, ticket_price, num_of_tickets, language, movie_3D):
        self.movieId = movieId
        self.title = title
        self.ticket_price = ticket_price
        self.num_of_tickets = num_of_tickets
        self.language = language
        self.movie_3D = movie_3D

    def calculate_total_price(self):
        return self.num_of_tickets * self.ticket_price

    def get_3d_charge(self):
      if self.movie_3D:
        return 50
      else:
        return 0

    def get_gst_rate(self):
        if self.language.lower() == 'hindi':
            return 8
        elif self.language.lower() == 'telugu':
            return 5
        elif self.language.lower() == 'english':
            return 10
        elif self.language.lower() == 'tamil':
            return 10
        else:
            return 5  # Default GST rate

    def calculate_gst(self, total_price):
        gst_rate = self.get_gst_rate()
        return total_price * (gst_rate / 100)

    def display_bill(self):
        total_price = self.calculate_total_price()
        gst_amount = self.calculate_gst(total_price)
        three_d_charge = self.get_3d_charge()
        final_bill = total_price + gst_amount + three_d_charge

        print(f"Movie ID: {self.movieId}")
        print(f"Movie Language: {self.language}")
        print(f"3D Movie       : {'Yes' if self.movie_3D else 'No'}")
        print(f"Title: {self.title}")
        print(f"Ticket Price: {self.ticket_price}")
        print(f"Number of Tickets: {self.num_of_tickets}")
        print(f"Total Price: {total_price}")
        if self.movie_3D:
            print(f"3D Charge: {three_d_charge}")
        print(f"GST Amount: {gst_amount}")
        print(f"Final Bill: {final_bill}")

cus1 = Movie(1, "LEO", 250, 2, "telugu", True)
cus1.display_bill()
```

```python
class User:
    def __init__(self, user_id, name, email, password):
        self.id = user_id
        self.name = name
        self.email = email
        self.password = password

    def login(self, email, password):
        return self.email == email and self.password == password


class Passenger:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Passenger Details")
        print("Name:", self.name)
        print("Age:", self.age)


class Flight:
    def __init__(self, flight_id, source, destination, capacity):
        self.id = flight_id
        self.source = source
        self.destination = destination
        self.capacity = capacity
        self.seats_booked = 0
        self.passengers = []

    def is_available(self, seats):
        return self.seats_booked + seats <= self.capacity

    def book_seats(self, seats):
        if self.is_available(seats):
            self.seats_booked += seats
            return True
        return False

    def display(self):
        print(f"\n************ Flight ID: {self.id} ************")
        print(f"Source: {self.source}")
        print(f"Destination: {self.destination}")
        print(f"Capacity: {self.capacity}")
        print(f"Seats Available: {self.capacity - self.seats_booked}")
        print("**********************************************")


class Booking:
    booking_counter = 1000

    def __init__(self, user, flight, seats):
        Booking.booking_counter += 1
        self.booking_id = Booking.booking_counter
        self.user = user
        self.flight = flight
        self.seats = seats
        self.passengers = []

        for i in range(seats):
            print(f"\nEnter Passenger {i+1} details")
            name = input("Name: ")
            age = int(input("Age: "))
            passenger = Passenger(name, age)
            self.passengers.append(passenger)

    def display(self):
        print(f"\n************ Booking ID: {self.booking_id} ************")
        print(f"Passenger Name: {self.user.name}")
        print(f"Flight ID: {self.flight.id}")
        print(f"Seats Booked: {self.seats}")
        print(f"From: {self.flight.source}")
        print(f"To: {self.flight.destination}")
        for passenger in self.passengers:
            passenger.display()
        print("**********************************************")


# Global data
flights = []
users = []
bookings = []
logged_in_user = None
flight_id_counter = 1000


def show_guest_menu():
    print("\n--- Airline Reservation System ---")
    print("1. Register")
    print("2. Login")
    print("3. Exit")
    choice = input("Choice: ")
    if choice == '1':
        register()
    elif choice == '2':
        login()
    elif choice == '3':
        print("Thank you for using the system.")
        exit()
    else:
        print("Invalid choice")


def register():
    name = input("Enter your name: ")
    email = input("Enter email: ")
    password = input("Enter password: ")
    user = User(len(users) + 1, name, email, password)
    users.append(user)
    print("Registered successfully.")


def login():
    global logged_in_user
    email = input("Enter email: ")
    password = input("Enter password: ")
    for user in users:
        if user.login(email, password):
            logged_in_user = user
            print(f"Login successful. Welcome, {user.name}!")
            return
    print("Login failed.")


def logout():
    global logged_in_user
    logged_in_user = None
    print("Logged out successfully.")


def show_user_menu():
    print(f"\n--- Welcome, {logged_in_user.name} ---")
    print("1. View Flights")
    print("2. Book Flight")
    print("3. View My Bookings")
    print("4. Logout")
    choice = input("Choice: ")
    if choice == '1':
        view_flights()
    elif choice == '2':
        book_flight()
    elif choice == '3':
        view_my_bookings()
    elif choice == '4':
        logout()
    else:
        print("Invalid choice")


def view_flights():
    print("\n--- Available Flights ---")
    for flight in flights:
        flight.display()


def book_flight():
    view_flights()
    flight_id = input("Enter Flight ID to book: ")
    seats = int(input("Enter number of seats: "))
    for flight in flights:
        if flight.id == flight_id:
            if flight.book_seats(seats):
                booking = Booking(logged_in_user, flight, seats)
                bookings.append(booking)
                print(f"Booking successful. Booking ID: {booking.booking_id}")
            else:
                print("Not enough seats available.")
            return
    print("Flight not found.")


def view_my_bookings():
    found = False
    for booking in bookings:
        if booking.user == logged_in_user:
            booking.display()
            found = True
    if not found:
        print("No bookings yet.")


# Initial setup
flights.append(Flight(f"S{flight_id_counter + 1}", "New Delhi", "Hong-Kong", 400))
flights.append(Flight(f"S{flight_id_counter + 2}", "Hyderabad", "Dubai", 350))
flights.append(Flight(f"S{flight_id_counter + 3}", "Chennai", "Tokyo", 450))


# Entry point
while True:
    if logged_in_user is None:
        show_guest_menu()
    else:
        show_user_menu()
```
